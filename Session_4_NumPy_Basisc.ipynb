```python
# Signal bruit√© (r√©aliste)

# ===============================================
# SIGNAL AVEC BRUIT (Simulation r√©aliste)
# ===============================================

# Signal pur
temps = np.linspace(0, 1, 1000)
signal_pur = np.sin(2 * np.pi * 5* temps)

# Ajout de bruit gaussien
niveau_bruit = 0.3
bruit = np.random.random(len(temps)) * niveau_bruit
signal_bruite = signal_pur + bruit

# _________________________________________________
#  CALCUL DU SNR Signal-to-Noise Ratio
# _________________________________________________

puissance_signal = np.mean(signal_pur **2)
puissance_bruit = np.mean(bruit**2)
snr_lineaire = puissance_signal / puissance_bruit
snr_db = 10 * np.log10(snr_lineaire)

# ________________________________________________
#    VISUALISATION
# ________________________________________________

fig, axes = plt.subplots(3, 1, figsize=(14, 9))

# Signal pur
axes[0].plot(temps[:200], signal_pur[:200], color='blue', linewidth=2)
axes[0].set_ylabel('Amplitude')
axes[0].set_title('Signal Pur (sans bruit)')
axes[0].grid(True, alpha=0.3)

# Bruit seul
axes[1].plot(temps[:200], bruit[:200], color='gray', linewidth=1)
axes[1].set_ylabel('Amplitude')
axes[1].set_title(f'Bruit Gaussien (œÉ = {niveau_bruit})')
axes[1].grid(True, alpha=0.3)

# Signal bruit√©
axes[2].plot(temps[:200], signal_bruite[:200], color='red', linewidth=1.5, alpha=0.8)
axes[2].set_ylabel('Amplitude')
axes[2].set_xlabel('Temps (secondes)')
axes[2].set_title(f'Signal Bruit√© (SNR= {snr_db:.2f} db)')
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("-"*60)
print("      ANALYSE DE BRUIT")
print("-"*60)
print(f"Puissance du signal : {puissance_signal:.4f}")
print(f"Puissance du bruit  : {puissance_bruit:.4f}")
print(f"SNR                 : {snr_db:.2f} db")
print("-"*60)
```

  **interpr√©tation SNR :**
SNR > dB : Exelent signal, bruit n√©gligeable
SNR 10-20 dB: Bon signal, bruit visible mais g√©rable
SNR 0-10 dB: signal d√©grad√©, traitement n√©cessaire
SNR < 0 dB: bruit plus fort que le signal, tr√®s probl√©matique
Application radar: Un √©cho lointain a souvent un SNR de 5-10 dB. 
**T√¢che:** le d√©tecter

```python
# Bibliot√®que DSP NumPy
# Fonctions am√©lior√©es avec NumPy

# ======================================
# BIBLIOTHEQUE DSP v2.0 - VERSION NUMPY
# Auteur: Arafan
# Date: 07/01/2026
# Lieu: Saint-P√©tersbourg, Russie
# ======================================

import numpy as np

def generer_signal_sinusoidal(frequence, duree, taux_ech=1000, amplitude=1, phase=0):
    """
    G√©n√®re un signal sinuso√Ødal.

    Param√®tres:
        frequence (float) : Fr√©quence en Hz 
        duree (float) : Dur√©e en sec
        taux_ech (int) : Taux d'√©chantillons en Hz
        amplitude (float): Amplitude en signal
        phase (float) : Phase initiale en radian

    Rturne :
        tuple : (temps, signal)
    """

    nombre_ech = int(taux_ech *  duree)
    temps = np.linspace(0, duree, nombre_ech)
    signal = amplitude * np.sin(2*np.pi*temps*frequence + phase)
    return  temps, signal
   
def ajouter_bruit(signal, snr_db):
    """
    Ajoute du bruit gaussien pour obtenir un SNR cible.

    Param√®tres :
        signal (ndarray): Signal d'entr√©s
        snr_dbl (float) : SNR d√©sir√© en dB

    Retourne
        tuple : (signal_bruite, bruit)
    """
    puissance_signal = np.mean(signal**2)
    snr_lineaire = 10**(snr_db / 10)
    puissance_bruit = puissance_signal / snr_lineaire
    
    bruit = np.random.randn(len(signal))*np.sqrt(puissance_bruit)
    signal_bruite = signal + bruit
    return signal_bruite, bruit

def calculer_snr(signal_pur, signal_bruite):
    """
    Calcule le rapport signal sur bruit.

    Param√®tres :
        signal_pur = (ndarray) : Signal original
        signal_bruite ndarray) : Signal avec bruit

    Retourne :
        float: SNR en d√©cibel
    """
    bruit = signal_bruite - signal_pur
    puissance_signal = np.mean(signal_pur**2)
    puissance_bruit = np.mean(bruit**2)

    if puissance_bruit==0:
        return float('inf')

    snr_db = 10*np.log10(puissance_signal / puissance_bruit)
    return snr_db

def detecter_pics(signal, seuil, distance_min = 1):
    """
    D√©tecte les pics dans un signal.

    Param√®tres:
        signal (ndarray) : Signal d'entr√©e 
        seuil (float) : Amplitude minimale pour d√©tecter un pic
        distance_min (int) : Distance minimale entre 2 pics en √©chantillons

    Retourne:
        ndarray : Indices des pics
    """
    # Simple d√©tection : valeurs au dessus du seuil
    au_dessus_seuil = signal > seuil

    # Trouver les indices
    indices = np.where(au_dessus_seuil)[0]

    if len(indices)==0:
        return np.array([])

    # Filtrer les pics trop proches
    pics = [indices[0]]
    for idx in indices [1:]:
        if idx - pics[-1] >= distance_min:
            pics.append(idx)
    return np.array(pics)

def normaliser_signal(signal):
    """
    Normaliser un signal entre -1 et 1.

    Param√®tres:
        signal (ndarray) : Signal d'entr√©e 

    Retourne :
        ndarray : Signal normalis√©
    """
    max_abs = np.max(np.abs(signal))
    if max_abs == 0:
        return signal
    return signal/max_abs

def energie_signal(signal) :
    """Calcule l'√©nergie totale d'un signal.

    Param√®tres:
        signal (ndarray): signal d'entr√©e 

    Retourne :
        float : Energie (Somme des carr√©s)
    """
    return np.sum(signal**2)

def puissance_signal(signal):
    """
    Calcule la puissance moyenne d'un signal

    Param√®tres :
        signal (ndarray) : Signal d'entr√©e

    Retourne :
        float : Puissance moyenne
    """
    return np.mean(signal**2)
    

# =============================================
# TESTS DE LA BIBLIOTHEQUE
# =============================================

print("‚úÖ Biblioth√®que DSP v2.0 (NumPy) charg√©")
print("\nüìö Fonctions disponibles :")
print(" - generer_signal_sinusoidal ()")
print(" - ajouter_bruit()")
print(" - calculer_snr()")
print(" - detecter_pics()")
print(" - normaliser_signal()")
print(" - energie_signal()")
print(" - puissance_signal()")
```
