```python
# Probl√®me : Code r√©p√©titif et p√©nible
temperature1 = [25,26,24,27,28]
temperature2 = [30,31,29,32,30]
temperature3 = [20,21,19,22,20]

# Calcul de la moyenne - REPETE 3 fois !
moy1 = sum(temperature1)/len(temperature1)
moy2 = sum(temperature2)/len(temperature2)
moy3 = sum(temperature3)/len(temperature3)

print(f"Moyenne capteur 1 : {moy1:.2f} ¬∞C")
print(f"Moyenne capteur 2 : {moy2:.2f} ¬∞C")
print(f"Moyenne capteur 3 : {moy3:.2f} ¬∞C")
```

```python
# Solution avec fonction
# Definition de la fonction ( une seule fois )
def calculer_moyenne(donnees):
    """
    Calcule la moyenne d'une liste de nombre.

    Param√®tres :
        donnees (list) : Liste de valeurs num√©riques
    returne :
        float : la moyenne des valeurs
    """
    moyenne = sum(donnees)/len(donnees)
    return moyenne

# Maintenant, utilisation simple
temperature1 = [25,26,24,27,28]
temperature2 = [30,31,29,32,30]
temperature3 = [20,21,19,22,20]

print(f"Moyenne capteur 1 : {moy1:.2f} ¬∞C")
print(f"Moyenne capteur 2 : {moy2:.2f} ¬∞C")
print(f"Moyenne capteur 3 : {moy3:.2f} ¬∞C")
```

```python
# Fonction avec plusieurs param√®tres
def generer_signal(amplitude, frequence, duree, echantillons):
    """
    G√©n√©rer un signal sinuso√Ødal.

    Param√®tres :
        amplitude (float) : Amplitude du signal
        frequence (float) : Fr√©quence en Hz
        duree (float) : Dur√©e en secondes
        echantillons (int) : Nombre d'√©chantillons

    Retourne : 
        List : Signal g√©n√©r√©
    """
    import math
    signal = []
    for i in range(echantillons):
        temps = (i / echantillons) * duree
        valeur = amplitude * math.sin(2 * math.pi * frequence * temps)
        signal.append(valeur)

    return signal

# Test de la fonction
mon_signal = generer_signal(amplitude=1.0, frequence=5, duree=1, echantillons=100)

print(f"Signal g√©n√©r√©   : {len(mon_signal)} √©chantillons")
print(f"Premi√®re valeur : {mon_signal[0]:.4f}")
print(f"Valeur max      : {max(mon_signal):.4f}")
print(f"Valeur min      : {min(mon_signal):.4f}")
```

```python
# Fonction qui retourne plusieurs valeurs
def analyser_signal_complet(signal):
    """
    Analyse complete d'un signal.

    Retourne : (moyenne, minimum, maximum, √©cart)
    """
    moyenne = sum(signal)/len(signal)
    minimum = min(signal)
    maximum = max(signal)
    ecart = maximum - minimum 

    return moyenne, minimum, maximum, ecart

# Utilisation
signal_test = [10,15,12,18,14,11,16,13]
moy, min_val, max_val, ecart = analyser_signal_complet(signal_test)

print("="*40)
print("RAPPORT D'ANALYSE SIGNAL")
print("="*40)
print(f"Moyenne       : {moy:.2f}")
print(f"Minimum       : {min_val}")
print(f"Maximum       : {max_val}")
print(f"Ecart (range) : {ecart}")
print("=" *40)
```

```python
# Fonction avec valeur par d√©faut 
def filtrer_signal(signal, seuil_min=-100, seuil_max=100):
    """
    Filtre un signal en √©liminant les valeurs hors limites.

    Param√®tres:
        signal (list) : Signal √† filtrer
        seuil_min (float) : Valeur minimale
        seuil_max (float) : Valeur maximale

    Retourne :
        List : Signal filtr√©
    """
    signal_filtre = []

    for valeur in signal:
        if seuil_min <= valeur <= seuil_max:
            signal_filtre.append(valeur)

        return signal_filtre
# Test avec signaux bruit√©s
signal_bruite = [25, 150, 26, -200, 24, 27, 300, 28]

# Utilisation 1 : Avec les seuils par d√©faut
resultat1 = filtrer_signal(signal_bruite)
print("Avec seuils par d√©faut :", resultat1)

# Utilisation 2 : Avec seuils personnalis√©s
resultat2 = filtrer_signal(signal_bruite, seuil_min=20, seuil_max=30)
print("Avec seuils 20 - 30 :", resultat2)
```

```python
# Biblioth√®que de fonctions DSP
 # =========================================
# MA BIBLIOTHEQUE PERSONNELLE v1.0
# Auteur : Arafan
# Date : 04.01.2026
# =========================================
def moyenne_signal(signal):
    """Calcule la moyenne d'un signal."""
    return sum(signal)/len(signal)

def puissance_signal(signal):
    """Calcule la puissance moyenne d'un signal."""
    return sum([x**2 for x in signal]) / len(signal)

def normaliser_signal(signal):
    """Normalyse un signale entre -1 et 1."""
    max_abs = max([abs(x) for x in signal])
    if max_abs == 0:
        return signal
    return [x / max_abs for x in signal]

def detecter_pics(signal, seuil):
    """
    D√©tecter les indices o√π le signal depasse un seuil.

    Retourne : Liste des indices des pics
    """
    pics = []
    for i, valeur in enumerate(signal):
        if valeur >seuil:
            pics.append(i)
        return pics

def rapport_signal_bruit(signal_pur, signal_bruite):
    """
    Calcule le rapport signal/bruit (SNR) en dB.
    """
    import math

    puissance_signal = puissance_signal(signal_pur)

    bruit = [signal_bruite[i] - signal_pur[i] for i in range(len(signal_pur))]
    puissance_bruit = puissance_signal(bruit)

    if puissance_bruit == 0:
        return float('inf')

    snr = 10 * math.log10(puissance_signal / puissance_bruit)
    return snr

# ======================
# TESTS DES FONCTIONS
# ======================

print("‚úÖBiblioth√®que DSP v1.0 chrg√©")
print("\nFonctions disponibles :")
print(" - moyenne_signal()")
print(" - puissance_signa()")
print(" - normaliser_signal()")
print(" - detecter_pics()")
print(" - rapport_signal_bruit()")
```

```python
# Analyseur de signal Radar Simplif√©
def simuler_echo_radar(distance, puissance_emission, attenuation=0.01):
    """
    Simuler les echo radar simple

    Param√®tres:
        distance(float) : Distance de la cible en m√®tre
        Puissance_emission (float) : Puissance √©mise en watts
        attenuation (float) : Coefficient d'att√©nuation

    Retourne :
        dict Information sur l'√©cho
    """
    import math

    # Calcul du temps aller-retour (vitesse lumi√®re = 3e8 m/s)
    temps_aller_retour = (2 * distance) / 3e8

    # Att√©nuation selon distance (loi inverse du carr√©)
    puissance_retour = puissance_emission / (distance **2 * attenuation)

    # D√©tectabilit√©  en watts
    seuil_detection = 0.0001 
    detecte = puissance_retour > seuil_detection

    return {
        'distance': distance,
        'temps_ms': temps_aller_retour * 1000, 
        'puissance_retour': puissance_retour,
        'detecte': detecte
    }

# Simulation de 5 cibles √† diff√©rentes distances
print("="*50)
print("SIMUTATION RADAR - DETECTION DE CIBLES")
print("="*50)
print("\nParam√®tres : Puissance √©mission = 1000 w\n")

distances_test = [100, 500, 1000, 5000, 10000]
for dist in distances_test:
    resultat = simuler_echo_radar(dist, puissance_emission=1000)

    statut = "üü¢DETECTER" if resultat['detecte'] else "üî¥ HORS PORTEE"

    print(f"Distance : {dist:5d} m | "
        f"Temps : {resultat['temps_ms']:.6f} ms | "
        f"Puissance retour : {resultat['puissance_retour']:.2f} w "
        f"{statut}")
print("="*50)
```
